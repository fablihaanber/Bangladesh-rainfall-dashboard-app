Dataset taken from https://climateserv.servirglobal.net/map

```python
import rasterio
import numpy as np
import pandas as pd
import geopandas as gpd
from rasterstats import zonal_stats
import os
from datetime import datetime
```

```python
# ── 1. Load Bangladesh district boundaries ──────────────────
districts = gpd.read_file("data/bgd_admin_boundaries.shp/bgd_admin2.shp")
print("Total districts:", len(districts))
print("Districts:", districts["adm2_name"].tolist())
```

```python
# ── 2. Set your CHIRPS folder path ──────────────────────────
folder = "data/rainfall_bangladesh/"
```

```python
# ── 3. Loop through all .tif files ──────────────────────────
results = []
files = sorted([f for f in os.listdir(folder) if f.endswith(".tif")])
total = len(files)

for idx, filename in enumerate(files):
    
    # Parse date from filename
    date = datetime.strptime(filename.replace(".tif", ""), "%Y%m%d")
    filepath = os.path.join(folder, filename)
    
    # Progress tracker
    print(f"Processing {filename} ({idx+1}/{total})...")
    
    try:
        # Zonal stats for all 64 districts at once
        stats = zonal_stats(
            districts,
            filepath,
            stats=["mean", "sum", "max", "min"],
            nodata=-9999
        )
        
        # Save one row per district
        for i, stat in enumerate(stats):
            results.append({
                "date":           date.strftime("%Y-%m-%d"),
                "year":           date.year,
                "month":          date.month,
                "day":            date.day,
                "district":       districts.iloc[i]["adm2_name"],
                "division":       districts.iloc[i]["adm1_name"],  # bonus!
                "mean_rainfall":  round(stat["mean"] or 0, 3),
                "total_rainfall": round(stat["sum"]  or 0, 3),
                "max_rainfall":   round(stat["max"]  or 0, 3),
                "min_rainfall":   round(stat["min"]  or 0, 3),
            })
    
    except Exception as e:
        print(f"  ⚠️ Error processing {filename}: {e}")
        continue

```

```python
# ── 4. Convert to DataFrame ──────────────────────────────────
df = pd.DataFrame(results)
```

```python
# ── 5. Preview ───────────────────────────────────────────────
print("\n=== PREVIEW ===")
print(df.head(10))
print("\nShape:", df.shape)
print("Columns:", df.columns.tolist())
print("Districts:", df["district"].nunique())
print("Date range:", df["date"].min(), "to", df["date"].max())
```
