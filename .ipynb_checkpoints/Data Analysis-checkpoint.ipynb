Yes, it's possible. Here are the steps:

**Data Source**
Download Bangladesh DEM data from one of these:
- **SRTM 30m** — [earthexplorer.usgs.gov](https://earthexplorer.usgs.gov) (free, most common)
- **ALOS DEM 12.5m** — [asf.alaska.edu](https://asf.alaska.edu) (higher resolution)
- **Copernicus DEM 30m** — [copernicus.eu](https://copernicus.eu) (newest, most accurate)

**Python Libraries needed**
```bash
pip install rasterio richdem matplotlib numpy
```

**Steps in code**

```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt
from rasterio.mask import mask
import geopandas as gpd

# 1. Load your shapefile to clip to Bangladesh boundary
bd_shape = gpd.read_file("bgd_admin_boundaries.shp/bgd_admin2.shp")
bd_union  = bd_shape.dissolve()

# 2. Load and clip DEM raster to Bangladesh
with rasterio.open("srtm_bangladesh.tif") as src:
    bd_union = bd_union.to_crs(src.crs)
    dem, transform = mask(src, bd_union.geometry, crop=True)
    dem = dem[0].astype(float)
    dem[dem == src.nodata] = np.nan

# 3. Plot
plt.figure(figsize=(10, 12))
plt.imshow(dem, cmap="terrain")
plt.colorbar(label="Elevation (m)")
plt.title("Bangladesh Digital Elevation Map")
plt.axis("off")
plt.savefig("bangladesh_dem.png", dpi=150, bbox_inches="tight")
```
